```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('Titanic')
df
```

```python
# scatterplot helps in establishing the relationship between two numerical data
```

```python
sns.scatterplot(x='age',y='fare',hue='sex',size='sex',data=df)
```

```python
sns.scatterplot(x='age',y='fare',hue='survived',size='sex',data=df)
```

```python
df2=sns.load_dataset('tips')
df2
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='sex',size='smoker',data=df2)
```

```python
# joinplot
#gives both the distribution along with the scatterplot
```

```python
sns.jointplot(x='total_bill',y='tip',kind='scatter',color='red',data=df2)
```

```python
sns.jointplot(x='total_bill',y='tip',kind='scatter',color='red',hue='sex',data=df2)
```

```python
sns.jointplot(x='total_bill',y='tip',color='red',data=df2,kind='reg')
```

```python
numeric_df = df2.select_dtypes(include=['number'])
correlation_matrix = numeric_df.corr()
```
