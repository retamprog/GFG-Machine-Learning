```python
def binary_search(lst,start,end,target):
    s=start
    l=end
    while s<=l:
        mid=s+(l-s)//2
        if lst[mid]==target:
            return mid
        elif lst[mid]>target:
            l=mid-1
        else:
            s=mid+1
            
binary_search(lst_2,0,len(lst_2)-1,max(lst_2))        
```
