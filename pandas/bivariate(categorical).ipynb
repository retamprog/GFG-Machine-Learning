```python
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
df.head()
```

```python
sns.countplot(x='survived',hue='class',data=df)
```

```python
x = pd.crosstab(df['survived'],df['class'])
x
```

```python
sns.heatmap(x, annot = True,fmt='.3f',square=True,cmap='coolwarm')
```

```python
lst=df['class'].unique()
classes=df['class']

for x in lst:
    c=0
    for i in classes:
        if x==i:
            c+=1
    print(f' {x} class has {c} people ')        
            
        
```
