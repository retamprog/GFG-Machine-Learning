```python
df[df.duplicated(keep = 'first')]
# when we appy the keep value first ,it will assign the last values as True for duplicates
```

```python
df[df.duplicated(keep = 'last')]
# if we assign the keep value as the last it will assign the first values as the duplicates
```

```python
df[df.duplicated(keep = False)]
# by default the keep value is First but if we set it to false then it will print all 
# the duplicated rows
```

```python
df1 = df.drop_duplicates()
# by default it will take keep value as first
#so it will drop the last values keeping the first values
df2 = df.drop_duplicates(keep = 'last')
# if we keep the keep value as last then it will drop the first values and 
# keep the last values
```
